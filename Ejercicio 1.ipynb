```python
class Auto:
  def __init__(self, numruedas, numasientos, placa, color):
    self.__numruedas = numruedas
    self.__numasientos = numasientos
    self.__placa = placa
    self.__color = color
 # -- SISTEMA 1
  def SetNumRuedas (self, numruedas):
    self.__numruedas = numruedas
  def GetNumRuedas (self):
    return self.__numruedas
  
  # -- SISTEMA 2
  def SetNumAsientos (self, numasientos):
    self.__numasientos = numasientos
  def GetNumAsientos (self):
    return self.__numasientos
  
  # -- SISTEMA 3
  def SetPlaca (self, placa):
    self.__placa = placa
  def Getplaca (self):
    return self.__placa
  
  # -- SISTEMA 4
  def SetColor (self, color):
    self.__color
  def GetColor (self):
    return self.__color
  
```

```python
class AutoCarreras(Auto):
  def __init__(self, numruedas, numasientos, placa, color, nrocambios, turbo):
    super().__init__(numruedas, numasientos, placa, color)
    self.__nrocambios = nrocambios
    self.__turbo = turbo

  #SISTEMA 1.1 
  def SetNroCambios(self, nrocambios):
    self.__nrocambios = nrocambios
  def GetNroCambios(self):
    return self.__nrocambios
  
  #SISTEMA 1.2
  def SetTurbo(self, Turbo):
    self.__turbo = turbo
  def GetTurbo(self):
    return self.__turbo
```
